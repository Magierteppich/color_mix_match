### Use MSE (Mean Squared error)

source: https://www.pyimagesearch.com/2014/09/15/python-compare-two-images/

```python
img_ready = ip.img_ready("../00_data/00_test_pic_full")
```

```python
imageA = cv2.cvtColor(img_ready[11], cv2.COLOR_BGR2GRAY)
imageB = cv2.cvtColor(img_ready[28], cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(imageA)
plt.show()
```

```python
plt.imshow(imageB)
plt.show()
```

```python
def mse(imageA, imageB):
    err = np.sum(((imageA.astype("float") - imageB.astype("float")) ** 2))
    err /= float(imageA.shape[0] * imageA.shape[1])
    return err
```

```python
mse(imageA, imageB)
```

## use Structural Similarity Index

source: https://scikit-image.org/docs/dev/auto_examples/transform/plot_ssim.html

```python
from skimage.metrics import structural_similarity as ssim
```

```python
def compare_images(imageA, imageB):
    # compute the mean squared error and structural similarity
    # index for the images
    m = mse(imageA, imageB)
    s = ssim(imageA, imageB)
    # setup the figure
    fig = plt.figure("x")
    plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))
    # show first image
    ax = fig.add_subplot(1, 2, 1)
    plt.imshow(imageA, cmap = plt.cm.gray)
    plt.axis("off")
    # show the second image
    ax = fig.add_subplot(1, 2, 2)
    plt.imshow(imageB, cmap = plt.cm.gray)
    plt.axis("off")
    # show the images
    plt.show()
```

```python
compare_images(imageA, imageB)
```

```python
s = ssim(imageA, imageB)
```

```python
# the higher the SSIM, the similar are the 2 images
# the lower the MSE, the similar are the 2 images
```

```python
valid_path, features, feature_list = imn.img_get_feature("../00_data/00_test_pic")
```

```python
feature_list
```

```python
valid_path
```

```python
target_index = imn.find_target_image(valid_path, target_image = "red")
target_index
```

```python
img_ready = imn.img_ready("../00_data/00_test_pic")
img_ready
```

```python
img_ready_gray = []
for img in img_ready: 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_ready_gray.append(gray)

#convert all img_ready into GRAY scale
```

```python
target_img_g = img_ready_gray[target_index]
img_mse = []

for img_g in img_ready_gray: 

    err = np.sum(((target_img_g.astype("float") - img_g.astype("float")) ** 2))
    err /= float(target_img_g.shape[0] * img_g.shape[1])
    img_mse.append([err]) 

#calcualate MSE based on GRAYscale image
img_mse
```

```python
img_ssim = []
for img_g in img_ready_gray: 
    similariy = ssim(target_img_g, img_g)
    img_ssim.append([similariy])
    
# calculated SSIM
img_ssim
```

```python
# 3. step - adding MSE and SSIM to the feature_list_knn

feature_list_knn = []
for i in range(len(valid_path)):
    temp = feature_list[i]+ img_mse[i] + img_ssim[i]
    feature_list_knn.append(temp)
```
